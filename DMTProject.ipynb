```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "BostonHousing.csv"
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(data.head())
```

```python
# Check the shape of the dataset
print("\nDataset shape:", data.shape)
```

```python
# Check for missing values
print("\nMissing values:")
print(data.isnull().sum())
```

```python
# Display data types and basic info
print("\nDataset information:")
print(data.info())
```

```python
# Summary statistics
print("\nSummary statistics:")
print(data.describe())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['medv'], kde=True, bins=30)
plt.title("Distribution of MEDV (Home Prices)")
plt.xlabel("MEDV ($1000s)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.pairplot(data[['medv', 'rm', 'lstat', 'crim', 'tax']])
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# RM vs MEDV
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['rm'], y=data['medv'])
plt.title("Number of Rooms vs Home Prices")
plt.xlabel("Average Number of Rooms (RM)")
plt.ylabel("Home Prices (MEDV)")
plt.show()

# LSTAT vs MEDV
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['lstat'], y=data['medv'])
plt.title("LSTAT vs Home Prices")
plt.xlabel("% Lower Status Population (LSTAT)")
plt.ylabel("Home Prices (MEDV)")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
data.boxplot(column=['crim', 'rm', 'lstat', 'tax'], vert=False)
plt.title("Boxplot of Key Features")
plt.show()
```

```python
data[['crim', 'rm', 'lstat', 'tax']].hist(bins=15, figsize=(12, 10), layout=(2, 2))
plt.suptitle("Distribution of Continuous Features")
plt.show()
```

```python
sns.countplot(x=data['chas'])
plt.title("Distribution of CHAS (Proximity to Charles River)")
plt.xlabel("CHAS (1 = near river, 0 = otherwise)")
plt.ylabel("Count")
plt.show()
```
